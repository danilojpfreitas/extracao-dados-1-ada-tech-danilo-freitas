Cria um consumidor e imprime os resultados a partir do tópico

```
import json
```

```
consumer = KafkaConsumer("primeiro_topico", bootstrap_servers=["localhost:9092"])

for message in consumer:
    mensagem_str = message.value.decode('utf-8')
    json_data = json.loads(mensagem_str)
    print(json_data)
```
